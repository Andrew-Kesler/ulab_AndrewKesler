```python
import matplotlib.pyplot as plt
import numpy as np

def plot_functions():
    """
    Function to plot f(x) = 2x + 1 and g(x) = x^2 - 3x + 2 on the same plot.
    Returns:
        f_values (list): A list of y values for f(x).
        g_values (list): A list of y values for g(x).
    """
    # Step 1: Generate x values using np.linspace for the domain [-10, 10]
    x_values = np.linspace(-10, 10, 100)
    
    # Step 2: Define the two functions
    f_values = 2 * x_values + 1
    g_values = x_values**2 - 3 * x_values + 2
    
    # Step 3: Create a new figure with a figsize of (10, 8)
    plt.figure(figsize=(10, 8))
    
    # Step 4: Plot f(x) with a solid blue line
    plt.plot(x_values, f_values, color='blue', linestyle='-', label='f(x) = 2x + 1')
    
    # Step 5: Plot g(x) with a dashed red line
    plt.plot(x_values, g_values, color='red', linestyle='--', label='g(x) = x^2 - 3x + 2')
    
    # Step 6: Add labels to the x and y axes
    plt.xlabel("X-axis", fontsize=12)
    plt.ylabel("Y-axis", fontsize=12)
    
    # Step 7: Add a title to the plot
    plt.title("Plot of f(x) and g(x)", fontsize=16)
    
    # Step 8: Add a legend to the plot
    plt.legend()
    
    # Step 9: Add a grid for better visualization
    plt.grid(True)
    
    # Step 10: Display the plot
    plt.show()
    
    # Return the lists of y values for both functions
    return f_values.tolist(), g_values.tolist()

# Call the function and print the returned lists
f_list, g_list = plot_functions()
print("f(x) values:", f_list)
print("g(x) values:", g_list)
```

```python
# To modify the function to also output the sum of the f(x) and g(x) values, you can follow these steps:

# 1. First, calculate the sum of the f(x) and g(x) values inside the function.
# You can use a list comprehension to add corresponding elements:
sum_values = [f + g for f, g in zip(f_values, g_values)]

# Explanation:
# - `zip(f_values, g_values)` pairs each element from `f_values` with the corresponding element from `g_values`.
# - The list comprehension `[f + g for f, g in zip(f_values, g_values)]` adds each pair together.
# - This results in a new list `sum_values` containing the sum of f(x) and g(x) for each x value.

# 2. Alternatively, you can use NumPy for element-wise addition:
sum_values = np.add(f_values, g_values)

# Explanation:
# - `np.add(f_values, g_values)` performs element-wise addition between the two NumPy arrays.
# - This method is more efficient and preferred for large datasets.

# 3. Finally, return the `sum_values` list along with the existing return values.
# Example code to include in your function:
return f_values.tolist(), g_values.tolist(), sum_values

# By returning `sum_values`, the function will output the combined values in addition to f(x) and g(x).
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_and_calculate_sums_v2():
    """
    Function to calculate f(x), g(x), and the sum of f(x) and g(x) over a given range of x values.
    Also plots the functions on a graph.
    Returns:
        f_values (list): List of f(x) values.
        g_values (list): List of g(x) values.
        sum_values_list (list): Sum of f(x) and g(x) using a list comprehension.
        sum_values_np (list): Sum of f(x) and g(x) using NumPy.
    """
    # Step 1: Generate x values using np.linspace for the domain [-10, 10]
    x_values = np.linspace(-10, 10, 100)
    
    # Step 2: Calculate f(x) and g(x)
    f_values = 2 * x_values + 1
    g_values = x_values**2 - 3 * x_values + 2
    
    # Step 3: Calculate the sum of f(x) and g(x) using a list comprehension
    sum_values_list = [f + g for f, g in zip(f_values, g_values)]
    
    # Step 4: Calculate the sum of f(x) and g(x) using NumPy
    sum_values_np = np.add(f_values, g_values)
    
    # Step 5: Plot the functions and their sum
    plt.figure(figsize=(10, 8))
    
    # Plot f(x)
    plt.plot(x_values, f_values, color='blue', linestyle='-', label='f(x) = 2x + 1')
    
    # Plot g(x)
    plt.plot(x_values, g_values, color='red', linestyle='--', label='g(x) = x^2 - 3x + 2')
    
    # Plot the sum of f(x) and g(x) using the list comprehension method
    plt.plot(x_values, sum_values_list, color='green', linestyle=':', label='Sum of f(x) + g(x)')
    
    # Add labels, title, and grid
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Plot of f(x), g(x), and their Sum")
    plt.grid(True)
    
    # Add a legend outside the plot
    plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
    
    # Display the plot
    plt.show()
    
    # Step 6: Return the lists of f(x), g(x), and their sum (both list comprehension and NumPy methods)
    return f_values.tolist(), g_values.tolist(), sum_values_list, sum_values_np.tolist()

# Call the function and show its output
f_list, g_list, sum_list, sum_np = plot_and_calculate_sums_v2()

# Print the outputs
print("f(x) values:", f_list)
print("g(x) values:", g_list)
print("Sum (using list comprehension):", sum_list)
print("Sum (using NumPy):", sum_np)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_horizontal_subplots(domain):
    """
    Function to create two side-by-side subplots:
    Left subplot: f(x) = 2x + 1
    Right subplot: g(x) = x^2 - 3x + 2

    Parameters:
        domain (tuple): A tuple representing the range of x values (min, max).
    """
    # Generate x values using np.linspace within the specified domain
    x_values = np.linspace(domain[0], domain[1], 100)
    
    # Define the functions
    f_values = 2 * x_values + 1
    g_values = x_values**2 - 3 * x_values + 2

    # Create a figure with two horizontal subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Left subplot for f(x)
    ax1.plot(x_values, f_values, color='blue')
    ax1.set_title("f(x) = 2x + 1")
    ax1.set_xlabel("X-axis")
    ax1.set_ylabel("Y-axis")
    ax1.grid(True)

    # Right subplot for g(x)
    ax2.plot(x_values, g_values, color='red')
    ax2.set_title("g(x) = x^2 - 3x + 2")
    ax2.set_xlabel("X-axis")
    ax2.set_ylabel("Y-axis")
    ax2.grid(True)

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()

def plot_vertical_subplots(domain):
    """
    Function to create two vertical subplots:
    Top subplot: f(x) = 2x + 1
    Bottom subplot: g(x) = x^2 - 3x + 2

    Parameters:
        domain (tuple): A tuple representing the range of x values (min, max).
    """
    # Generate x values using np.linspace within the specified domain
    x_values = np.linspace(domain[0], domain[1], 100)
    
    # Define the functions
    f_values = 2 * x_values + 1
    g_values = x_values**2 - 3 * x_values + 2

    # Create a figure with two vertical subplots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

    # Top subplot for f(x)
    ax1.plot(x_values, f_values, color='blue')
    ax1.set_title("f(x) = 2x + 1")
    ax1.set_xlabel("X-axis")
    ax1.set_ylabel("Y-axis")
    ax1.grid(True)

    # Bottom subplot for g(x)
    a2.plot(x_values, g_values, color='red')
    ax2.set_title("g(x) = x^2 - 3x + 2")
    ax2.set_xlabel("X-axis")
    ax2.set_ylabel("Y-axis")
    ax2.grid(True)

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_surface():
    """
    Function to create a 3D surface plot.
    """
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Generate x and y data
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    x, y = np.meshgrid(x, y)
    
    # Define the z function
    z = np.sin(np.sqrt(x**2 + y**2))
    
    # Plot the surface
    surface = ax.plot_surface(x, y, z, cmap='viridis')
    ax.set_title("3D Surface Plot")
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_zlabel("Z-axis")
    fig.colorbar(surface)
    
    plt.show()

def plot_3x3_grid():
    """
    Function to create a 3x3 grid of subplots.
    """
    fig, axes = plt.subplots(3, 3, figsize=(12, 12))
    
    # Generate x values
    x = np.linspace(0, 10, 100)
    
    for i in range(3):
        for j in range(3):
            # Define a unique function for each subplot
            y = np.sin(x + i) + np.cos(x + j)
            axes[i, j].plot(x, y)
            axes[i, j].set_title(f"Plot ({i+1},{j+1})")
            axes[i, j].set_xlabel("X-axis")
            axes[i, j].set_ylabel("Y-axis")
            axes[i, j].grid(True)
    
    plt.tight_layout()
    plt.show()

def plot_colormaps():
    """
    Function to visualize an image with different colormaps.
    """
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    
    # Generate random data
    data = np.random.rand(10, 10)
    
    # Define a list of colormaps
    colormaps = ['viridis', 'plasma', 'inferno', 'cividis']
    
    # Plot the data with different colormaps
    for ax, cmap in zip(axes.flat, colormaps):
        im =ax.imshow(data, cmap=cmap)
        ax.set_title(f"Colormap: {cmap}")
        fig.colorbar(im, ax=ax)
    
    plt.tight_layout()
    plt.show()
```
